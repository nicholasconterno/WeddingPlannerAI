```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

def send_email(sender_email, sender_password, recipient_email, subject, body):
    try:
        # Create message container
        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = recipient_email
        msg['Subject'] = subject

        # Attach body to the message
        msg.attach(MIMEText(body, 'plain'))

        # Setup the SMTP server
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(sender_email, sender_password)

        # Send the email
        server.sendmail(sender_email, recipient_email, msg.as_string())
        server.quit()
        print("Email sent successfully!")
    except smtplib.SMTPAuthenticationError:
        print("Failed to authenticate with the SMTP server. Check your email and password.")
    except Exception as e:
        print(f"An error occurred: {e}")


# Example usage
send_email('nicholas.e.conterno@gmail.com', 'jidw sjei rkkn isbz', 'nconterno@me.com', 'Test Subject', 'This is a test email')
```

```python
import imaplib
import email

def receive_email(email_user, email_app_password):
    try:
        # Connect to the mail server
        mail = imaplib.IMAP4_SSL('imap.gmail.com')
        mail.login(email_user, email_app_password)
        mail.select('inbox')

        # Search for all emails in the inbox
        status, data = mail.search(None, 'ALL')
        mail_ids = data[0].split()

        # Fetch the latest email
        latest_email_id = mail_ids[-1]
        status, data = mail.fetch(latest_email_id, '(RFC822)')

        # Parse the email
        raw_email = data[0][1]
        msg = email.message_from_bytes(raw_email)

        # Print email details
        print('From:', msg['From'])
        print('Subject:', msg['Subject'])

        if msg.is_multipart():
            for part in msg.walk():
                content_type = part.get_content_type()
                if content_type == 'text/plain':
                    print('Body:', part.get_payload(decode=True).decode())
        else:
            print('Body:', msg.get_payload(decode=True).decode())

        mail.logout()
        print("Email received successfully!")
    except imaplib.IMAP4.error:
        print("Failed to authenticate with the IMAP server. Check your email and app password.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
receive_email('nicholas.e.conterno@gmail.com', 'jidw sjei rkkn isbz')
```

```python
import imaplib
import email

def receive_emails(email_user, email_app_password, max_emails=10):
    try:
        # Connect to the mail server
        mail = imaplib.IMAP4_SSL('imap.gmail.com')
        mail.login(email_user, email_app_password)
        mail.select('inbox')

        # Search for all emails in the inbox
        status, data = mail.search(None, 'ALL')
        mail_ids = data[0].split()

        # Fetch the latest emails up to max_emails
        for i in range(min(len(mail_ids), max_emails)):
            latest_email_id = mail_ids[-(i+1)]
            status, data = mail.fetch(latest_email_id, '(RFC822)')

            # Parse the email
            raw_email = data[0][1]
            msg = email.message_from_bytes(raw_email)

            # Print email details
            print(f"Email {i+1}:")
            print('From:', msg['From'])
            print('Subject:', msg['Subject'])

            if msg.is_multipart():
                for part in msg.walk():
                    content_type = part.get_content_type()
                    if content_type == 'text/plain':
                        print('Body:', part.get_payload(decode=True).decode())
            else:
                print('Body:', msg.get_payload(decode=True).decode())
            print("="*50)

        mail.logout()
        print("Emails received successfully!")
    except imaplib.IMAP4.error:
        print("Failed to authenticate with the IMAP server. Check your email and app password.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
receive_emails('nicholas.e.conterno@gmail.com', 'jidw sjei rkkn isbz')
```
